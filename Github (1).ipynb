Use the Notebook to build the code to scrape the following Wikipedia page, https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M, in order to obtain the data that is in the table of postal codes and to transform the data into a pandas dataframe like the one shown below:

```python
import pandas as pd
import numpy as np
```

```python
tables = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M",header=0)[0]
tables
```

```python
df=pd.DataFrame(tables)
df.head()
```

```python
df.shape
```

```python
df['Borough'].replace("Not assigned", np.nan, inplace = True)
```

```python
df.dropna(subset=["Borough"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.shape
```

```python
df.columns=(['PostalCode','Borough','Neighborhood'])
```

```python
df.head(10)
```

```python
df.loc[df['Neighborhood'] == 'Not assigned', ['Neighborhood']] = df['Borough']
df.head(10)
```

```python
df=df.groupby(['PostalCode','Borough'], axis=0)['Neighborhood'].apply(','.join).reset_index()
df.head(10)
```

```python
df.shape
```

```python
df.to_csv('PostalCodes_peer3.csv',index=False)
```
